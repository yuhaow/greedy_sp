```python
%matplotlib inline

import matplotlib.pyplot as plt

import numpy as np
from scipy.stats import mannwhitneyu
```

```python
log_pvals = np.loadtxt('log_pvals')
genes = np.genfromtxt('genes', dtype='str')
perturbations = np.genfromtxt('perturbations', dtype='str')
```

```python
import matplotlib

font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 16}

matplotlib.rc('font', **font)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(6,6)
index = np.array(perturbations) != 'MouseNTC_100_A_670'
heatmap = plt.pcolor(log_pvals[:,index], cmap = 'coolwarm', vmin = -19, vmax = 19)
ax.set_yticks(np.arange(len(genes)) + 0.5, minor=False)
ax.set_xticks(np.arange(len(np.array(perturbations)[index])) + 0.5, minor=False)
ax.invert_yaxis()
ax.set_xticklabels(np.array(perturbations)[index], minor=False)
ax.set_yticklabels(genes, minor=False)
ax.set_xlabel('Deleted Genes')
ax.set_ylabel('Measured Genes')
plt.xticks(rotation=90)
plt.colorbar(label ='q-value')
```
